**Task 1 (easy): Solution single-variate functions**
Select the option(s) \(A, B,\) or \(C\) for which the input variables \((x, y, z, t)\) satisfy the single-variate functions.
1. A: \(x=10\), B: \(x=12\), C: \(x=15\)

\[3x-7=2x+5\]

2. A: \(y=1\), B: \(y=2\), C: \(y=3\)

\[4y+\tfrac y2=9\]

3. A: \(x=4\), B: \(x=5\), C: \(x=6\)

\[5(x-3)+2x=20\]

4. A: \(t=6\), B: \(t=4\), C: \(t=10\)

\[2t+3=\tfrac t2+9\]



---

```python
import numpy as np

a = np.array([[3-2]])
b = np.array([5+7])
x = np.linalg.solve(a, b)
print("x =", x[0])


a = np.array([[4 + 0.5]])  
b = np.array([9])

y = np.linalg.solve(a, b)
print("y =", y[0])  


a = np.array([[5 + 2]])  
b = np.array([35])

x = np.linalg.solve(a, b)
print("x =", x[0])  

a = np.array([[2 - 0.5]])  
b = np.array([6])

t = np.linalg.solve(a, b)
print("t =", t[0])  
```

**Task 2 (easy): Solution multi-variate functions**
Select the option(s) \(A, B,\) and/or \(C\) for which the input variables \((x_1, x_2, x_3)\) satisfy the multi-variate functions.
1. A: \((x_1=3, x_2=4)\), B: \((x_1=2, x_2=6)\), C: \((x_1=4, x_2=2)\)
\[2x_1+3x_2=18\]

2. A: \((x_2=2, x_1=7)\), B: \((x_1=6, x_2=1)\), C: \((x_2=4, x_1=8)\)
\[x_1-x_2=5\]

3. A: \((x_ 1=3, x_ 2=4)\), B: \((x_2=0, x_1=5)\), C: \((x_2=2, x_1=6)\)
\[x_1^2+x_2^2=25\]

4. A: \((x_1=2, x_2=4,, x_3=6)\), B: \((x_2=3, x_1=3, x_3=5)\), C: \((x_3=5, x_1=5, x_2=5)\)
\[x_1+x_2+x_3=12\]

5. A: \((x_3=3, x_1=2, x_2=1)\), B: \((x_2=2, x_1=3, x_3=3)\), C: \((x_1=4, x_3=5, x_2=1)\)
\[2x_1-x_2+x_3=7\]

6. A: \((x_2=5, x_1=2, x_3=4)\), B: \((x_3=2, x_2=4, x_1=3)\), C: \((x_1=4, x_2=3, x_3=2)\)
\[x_1^2+x_2+x_3=15\]

7. A: \((x_2=3, x_1=2, x_3=4)\), B: \((x_1=3, x_3=5, x_2=2)\), C: \((x_2=5, x_1=1, x_3=5)\)
\[x_1x_2+x_3=10\]



---

```python
import numpy as np

def linearEquation(coeffs,options,target):
    
    results = options @ coeffs
    matches = results == target

    index = 0
    for i in matches:
        index += 1
        if i:
            print(f"Option {index} matches")
            
# --------------------------           
print("The answer for 1:")
coeffs = np.array([2, 3])
options = np.array([
        [3, 4],  
        [2, 6],  
        [4, 2]   
    ])
target = 18
linearEquation(coeffs,options,target)
# --------------------------
print("The answer for 2:")
coeffs = np.array([1, -1])
options = np.array([
        [2, 7],  
        [6, 1],  
        [4, 8]   
    ])
target = 5
linearEquation(coeffs,options,target)
# --------------------------
coeffs = np.array([1,1])
opt = np.array([
        [3, 4],  
        [0, 5],  
        [2, 6]   
    ])
target = 25
options = opt*opt
print("The answer for 3:")
linearEquation(coeffs,options,target)
# --------------------------
print("The answer for 5:")
coeffs = np.array([2,-1,1])
options = np.array([
        [3,2,1],  
        [2,3,3],  
        [4,5,1]   
    ])
target = 7
linearEquation(coeffs,options,target)
print("None")

# --------------------------
coeffs = np.array([1,1,1])
opt = np.array([
        [25,2,4],  
        [4,4,3],  
        [16,3,2]   
    ])
target = 15
print("The answer for 6:")
linearEquation(coeffs,opt,target)
print("None")

# --------------------------
coeffs = np.array([1,1])
opt = np.array([
        [3*2,4],  
        [3*5,2],  
        [5*1,5]   
    ])
target = 10
print("The answer for 7:")
linearEquation(coeffs,opt,target)
```

**Task 3 (easy): Determine weights 1**
Given the affine model: 
\[y = w_1 x_1 + w_2 x_2 + w_3\]
Determine which of the weight options $(w_1, w_2, w_3)$ contain all data points below, such that all three points are on the plane:
**Data points:**
Input $x_i$, label $y$:
- $x_1 = 2, x_2 = 0, y = 5  $
- $x_1 = 1, x_2 = 1, y= 6  $
- $x_1 = 0, x_2 = 2, y= 7  $

**Options:**
1. $(w_1, w_2, w_3) = (1, 0, 3)$
2. $(w_1, w_2, w_3) = (1, 3, 2)$
3. $(w_1, w_2, w_3) = (2, 3, 1)$


---

```python
#Write your answer here...


def linearEquation(coeffs,values,y):
    
    results = values @ coeffs.T
    matches = results.T == y
    
    return matches
# ---------------------------------


coeffs = np.array([
        [1,0,3], 
        [1,3,2],  
        [2,3,1]   
])
values = np.array([
        [2,0,1],  
        [1,1,1],  
        [0,2,1]   
])

y = np.array([5, 6, 7])

linearEquation(coeffs,values,y)
   
            
```

**Task 5 (medium): Determine weights 2📽️**
Rewrite the affine model as an inner-product seperating weights and inputs. If not possible, explain why. 

A: 
$$y = w_1 \cos(x_1) + w_2 x_2^2 + w_3$$


B: 
$$y = e^{x_1} w_1  + w_2 x_2^2 + w_3$$
 
C: 
$$y =  w_3 + \sin(x_1) w_1  + w_2 \ln(x_2)$$


D: 
$$y =  w_1 x_1 x_2  + w_2 x_2 + w_3$$


E: 
$$y =  w_1 x_1 + w_2 \sin(x_2)  + w_3 x_3 + w_4 \cos(x_4) + w_5$$


F: 
$$y =  w_1 x_1 + w_4\sin(x_2)  + w_2 x_3 + w_3 \cos(x_4) + w_4$$


---

```python

# Inner-product possible: A, B, C, D, E
# Not possible: F
```


### A

**Inner-product form:**
$$
\mathbf{x} = 
\begin{bmatrix} \cos(x_1) \\ x_2^2 \\ 1 \end{bmatrix}, \quad
\mathbf{w} = 
\begin{bmatrix} w_1 \\ w_2 \\ w_3 \end{bmatrix}, \quad
y = \mathbf{w}^\top \mathbf{x}
$$

---

### B

**Inner-product form:**
$$
\mathbf{x} = 
\begin{bmatrix} e^{x_1} \\ x_2^2 \\ 1 \end{bmatrix}, \quad
\mathbf{w} = 
\begin{bmatrix} w_1 \\ w_2 \\ w_3 \end{bmatrix}, \quad
y = \mathbf{w}^\top \mathbf{x}
$$

---

### C

**Inner-product form:**
$$
\mathbf{x} = 
\begin{bmatrix} \sin(x_1) \\ \ln(x_2) \\ 1 \end{bmatrix}, \quad
\mathbf{w} = 
\begin{bmatrix} w_1 \\ w_2 \\ w_3 \end{bmatrix}, \quad
y = \mathbf{w}^\top \mathbf{x}
$$

---

### D


**Inner-product form:**
$$
\mathbf{x} = 
\begin{bmatrix} x_1 x_2 \\ x_2 \\ 1 \end{bmatrix}, \quad
\mathbf{w} = 
\begin{bmatrix} w_1 \\ w_2 \\ w_3 \end{bmatrix}, \quad
y = \mathbf{w}^\top \mathbf{x}
$$

---

### E

**Inner-product form:**
$$
\mathbf{x} = 
\begin{bmatrix} x_1 \\ \sin(x_2) \\ x_3 \\ \cos(x_4) \\ 1 \end{bmatrix}, \quad
\mathbf{w} = 
\begin{bmatrix} w_1 \\ w_2 \\ w_3 \\ w_4 \\ w_5 \end{bmatrix}, \quad
y = \mathbf{w}^\top \mathbf{x}
$$

---

### F
$$
y = w_1 x_1 + w_4 \sin(x_2) + w_2 x_3 + w_3 \cos(x_4) + w_4
$$

**Not possible as an inner-product**, because $w_4$ appears **twice** (once multiplied by $\sin(x_2)$ and once as a constant). In standard inner-product form each weight appears **only once**.
